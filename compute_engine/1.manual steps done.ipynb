## 🐳 **Basic Docker Commands (VM level)**

### 🔹 1. Check Docker installation

```bash
docker --version
docker info
```

---

### 🔹 2. See all **running containers**

```bash
docker ps
```

✅ Example output:

```
CONTAINER ID   IMAGE            COMMAND                  STATUS         PORTS               NAMES
a1b2c3d4e5f6   node:18          "npm start"              Up 3 hours     0.0.0.0:8080->80/tcp   backend_app
```

---

### 🔹 3. See **all containers** (including stopped ones)

```bash
docker ps -a
```

---

### 🔹 4. See **all images** in the VM

```bash
docker images
```

✅ Example output:

```
REPOSITORY    TAG       IMAGE ID       CREATED         SIZE
node          18        3c7a0f52c1c8   2 weeks ago     914MB
backend_app   latest    98b3c1d2e5aa   1 day ago       600MB
```

---

### 🔹 5. See **container logs**

```bash
docker logs <container_name_or_id>
```

Example:

```bash
docker logs backend_app
```

---

### 🔹 6. Enter a running container (get inside its shell)

```bash
docker exec -it <container_name_or_id> /bin/bash
```

If `/bin/bash` doesn’t work:

```bash
docker exec -it <container_name_or_id> sh
```

---

### 🔹 7. Stop / Start / Restart a container

```bash
docker stop <container_name_or_id>
docker start <container_name_or_id>
docker restart <container_name_or_id>
```

---

### 🔹 8. Remove a container

```bash
docker rm <container_name_or_id>
```

---

### 🔹 9. Remove an image

```bash
docker rmi <image_name_or_id>
```

---

### 🔹 10. Check **resource usage**

```bash
docker stats
```

Shows live CPU, memory, network usage per container.

---

## ⚙️ **Docker Compose (if used)**

If your backend is using **docker-compose.yml**, you can manage services like this:

```bash
docker compose ps          # List running services
docker compose up -d       # Start in detached mode
docker compose down        # Stop and remove containers
docker compose logs -f     # Follow logs
```

> (Older versions use `docker-compose` with a hyphen)

## 🔍 Helpful inspection commands

| Task                       | Command                           |
| -------------------------- | --------------------------------- |
| Inspect a container        | `docker inspect <container_name>` |
| See ports used             | `docker port <container_name>`    |
| Container resource usage   | `docker stats`                    |
| Cleanup stopped containers | `docker container prune`          |
| Cleanup unused images      | `docker image prune`              |
| Cleanup all unused data    | `docker system prune -a`          |

---

### 💡 Bonus: Find your backend container quickly

If you’re unsure which container is your backend:

```bash
docker ps --format "table {{.Names}}\t{{.Image}}\t{{.Ports}}\t{{.Status}}"
```

This gives a clean view of all running containers with their port mappings.
